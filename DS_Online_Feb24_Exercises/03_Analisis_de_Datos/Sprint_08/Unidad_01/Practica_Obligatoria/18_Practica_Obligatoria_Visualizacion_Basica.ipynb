### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
ciudades=['LosAngeles', 'SanDiego', 'Sacramento', 'Anaheim', 'SantaAna']
df_ciudades=df[df['city'].isin(ciudades)]
df_ciudades.sort_values(by= 'population_total', ascending= False, inplace= True)
df_ciudades
```

```python
'''
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). 
(Colores: azul, gris, aceituna, marrón, con trasparencia)
'''
plt.ticklabel_format(style='plain', axis='y')

plt.bar(df_ciudades['city'], df_ciudades['population_total']/1000, color= ['blue', 'grey', 'red','olive', 'brown'], 
        alpha= 0.5)
plt.xlabel('Ciudades')
plt.ylabel('Habitantes (miles)')
plt.ylim(0, 5000)
plt.title('Ciudades mas pobladas de California', fontsize= 10)
for i, valor in enumerate(df_ciudades['population_total']/1000):
    plt.text(i, valor + 1, str(valor), ha='center', va='bottom')
```

```python
'''Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)'''
plt.ticklabel_format(style='plain', axis='y')
font_dict = {
    'size':20,
    'family': "Bradley Hand itc",
}
plt.hist(df['area_land_sq_mi'], bins= 50, alpha= 0.7, color= 'steelblue')
plt.xlabel('Superficie ($km^2$)', fontdict= font_dict)
plt.ylabel('Numero de ciudades', fontdict= font_dict)
plt.legend(['Hist Area'], fontsize= 9);
```

```python
'''3. El scatter de área de agua de un ciudad con su poblacion.'''
arrowprops = {
    'arrowstyle': '->'
}

x_arrow_1 = df['area_water_km2'].max() - 73
y_arrow_1 =(df.set_index('population_total')['area_water_km2'].idxmax() - 3300000)/1000

x_arrow_2 =df.set_index('area_water_km2')['population_total'].idxmax() - 73
y_arrow_2 = (df['population_total'].max() - 3300000)/1000

plt.ticklabel_format(style='plain', axis='y')
plt.scatter(df['area_water_km2'], df['population_total']/1000, 
            alpha= 0.3, color= 'g')
plt.xlabel('Area de agua en la ciudad (km$^2$)')
plt.ylabel('Poblacion (miles de habitantes)')
plt.title('Poblacion por kilometro cuadrado de agua', fontsize= 10)
plt.text(60, 3300000/1000, 'Outliers', fontsize= 8)
plt.arrow(73, 3300000/1000, dx= x_arrow_1, dy= y_arrow_1)
plt.arrow(73, 3300000/1000, dx= x_arrow_2, dy= y_arrow_2)
```

```python
'''El boxplot de la altura de las ciudades.'''

boxprops = {
    'color': 'black',
    'facecolor': 'red'
}
medianprops = {
    'color': 'black'
}
plt.boxplot(df['elevation_m'], whis= 1.5, vert= False, boxprops= boxprops, 
            medianprops= medianprops, patch_artist= True)
plt.title('Diagrama de caja de la altitud de las ciudades de california', fontsize= 8)
plt.xlabel('Altitud (m)', fontsize= 8);
```
